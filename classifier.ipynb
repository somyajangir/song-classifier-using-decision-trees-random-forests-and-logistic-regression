```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
data1 = pd.read_json('echonest-metrics.json')
```

```python
data2 = pd.read_csv('fma-rock-vs-hiphop.csv')
```

```python
data2.head()
```

```python
# from above we can take genre top and merge to previous data
data = data1.merge(data2[['genre_top', 'track_id']], on='track_id')
```

```python
data.head()
```

```python
# Checking the co-relation to remove non-important labels
data.corr(numeric_only= True)
```

```python
#track_id is not relating with any data so drop
data.drop('track_id', axis =1, inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('genre_top', axis =1)
y = data['genre_top']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
#Giving 20% of data for testing and rest for training
```

```python
# Using Deision trees for classification
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
```

```python
#using random forest for more accuracy
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=20)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, rfc_pred))
print('\n')
print(classification_report(y_test, rfc_pred))
```

```python
# Random Forest fetch better result than decision trees here
```

```python
#Now , let us logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
```

```python
# So in this exaple we can see that logistic regression is fetching the best results.
```
